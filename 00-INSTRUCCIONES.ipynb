# **1.Introducción**



Home Credit es una empresa de filipinas que busca ofrecer créditos a personas con antecedentes crediticios insuficientes o inexistentes, buscando la inclusión financiera de la población no bancarizada.

Para ello, esta compañía busca utilizar información transaccional y de telecomunicaciones, con el fin de predecir el riesgo de default de sus clientes.

**Principal Objetivo**: ¿Al cliente, le debo o no de otorgar el préstamo a partir de su riesgo de default? 

**Variable Objetivo**: Probabilidad de que el cliente presente retrasos en el pago de su cuota  (1) o no (0)

En este proyecto se realizaron 8 iteraciones diferentes, correspondiente a los 8 notebooks los cuales su nombre inicia con "Modelacion_corrida..."



*   Iteración 1: 03-Modelacion_corrida1.ipynb

*   Iteración 3: 08_Modelacion_corrida_62cols.ipynb

*   Iteración 4: 10-Modelacion_corrida_60Cols.ipynb

*   Iteración 5: 11-Modelacion_corrida_60Cols_scaler.ipynb

*   Iteración 6: 12-Modelacion_corrida2_15Cols_imbalance.ipynb

*   Iteración 7: 13-Modelacion_corrida2_15Cols_balanced_over.ipynb

### **3.01.2.Recomendaciones para la ejecución**

En este primer notebook es importante tener presente varias observaciones:

1. Se debe hacer la conexión entre la api de kaggle y colab, para esto es necesario descargar el Token que genera kaggle. Este se genera como un archivo llamado **kaggle.json**, el cual debe ser cargado en colab, en la línea con código: **files.upload()**. (https://www.youtube.com/watch?v=T8xEQI8XXGs en este video muestran como generar el kaggle.json)

2. En la línea de código: **!kaggle competitions download -c home-credit-default-risk**, se hace la descarga del archivo o archivos, estos se descargan en uno o varios archivos .zip, por tanto al correr esta línea, hay que percatarse que el o los archivos que se descargan si sean los correspondiente a los que se descomprimen en la línea de código siguiente: ***! unzip home-credit-default-risk.zip ***, sino, debe cambiarse el nombre por el o los .zip que aparezcan en colab.
